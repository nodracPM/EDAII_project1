```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
tam = [50, 100, 500, 800, 1000, 2000, 5000, 10000]
bubble_sort = [1825.2, 7507.8, 185730.6, 478678.8, 744657.8, 2986262.2, 1.87e14, 7.47e15]
insertion_sort = [1300.0, 5224.8, 122828.0, 320448.8, 492108.0, 1980135.2, 1.24e15, 4.95e14]
selection_sort = [1319.2, 5141.6, 125742.2, 321185.6, 501483.0, 2002975.0, 1.25e15, 5.00e15]
pancake_sort = [2123.6, 8460.0, 205936.4, 524429.4, 818961.8, 3260795.4, 2.03e15, 8.13e15]

# Crear la gráfica
plt.figure(figsize=(10, 6))

# Graficar cada algoritmo
plt.plot(tam, bubble_sort, label="Bubble Sort", marker='o')
plt.plot(tam, insertion_sort, label="Insertion Sort", marker='o')
plt.plot(tam, selection_sort, label="Selection Sort", marker='o')
plt.plot(tam, pancake_sort, label="Pancake Sort", marker='o')

# Configurar escala logarítmica para el eje Y
plt.yscale('log')

# Añadir título y etiquetas
plt.title('Comparación de Algoritmos de Ordenamiento por Operaciones')
plt.xlabel('Tamaño del Arreglo')
plt.ylabel('Operaciones')
plt.grid(True)

# Añadir leyenda
plt.legend()

# Mostrar la gráfica
plt.show()

# Guardar la gráfica

plt.savefig('grafica_algoritmos_lentos.png', dpi=300) 
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
tamanos = [50, 100, 500, 800, 1000, 2000, 5000, 10000]
heap_sort = [1042.2, 2473.2, 16849.2, 29127.2, 37611.2, 83128.8, 234336.8, 507779.2]
quick_sort = [73.6, 151.8, 790.0, 1193.6, 1380.4, 3241.0, 7101.4, 17537.2]
merge_sort = [557.6, 1313.8, 8836.0, 15298.0, 19674.4, 43341.4, 121940.0, 263856.6]

# Crear la gráfica
plt.figure(figsize=(10, 6))

# Graficar cada algoritmo
plt.plot(tamanos, heap_sort, label="Heap Sort", color = 'purple', linewidth = 2, linestyle = '--', marker='o')
plt.plot(tamanos, quick_sort, label="Quick Sort", color = 'navy', linewidth = 2, linestyle = '--', marker='o')
plt.plot(tamanos, merge_sort, label="Merge Sort", color = 'yellow', linewidth = 2, linestyle = '--', marker='o')

# Configurar escala logarítmica para el eje Y
plt.yscale('log')

# Añadir título y etiquetas
plt.title('Comparación de Algoritmos de Ordenamiento por Operaciones')
plt.xlabel('Tamaño del Arreglo')
plt.ylabel('Operaciones')
plt.grid(True)

# Añadir leyenda
plt.legend()

# Mostrar la gráfica
plt.show()

# Guardar la gráfica

plt.savefig('grafica_algoritmos_rapidos.png', dpi=300) 
```
